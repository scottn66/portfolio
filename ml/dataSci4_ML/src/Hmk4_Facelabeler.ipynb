```python
from sklearn.datasets import fetch_lfw_people

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.decomposition import PCA
```

```python
lfw_people = fetch_lfw_people(min_faces_per_person=70, resize=0.4)
```

```python
# introspect the images arrays to find the shapes (for plotting)
n_samples, h, w = lfw_people.images.shape

X = lfw_people.data
n_features = X.shape[1]

# the label to predict is the id of the person
y = lfw_people.target
target_names = lfw_people.target_names
n_classes = target_names.shape[0]
```

```python
# Split into training set and testing set. Then compute a PCA on the dataset, 
#followed by a classification technique.
```

```python
print(X.shape)
print("features: " + str(n_features))
print(y)
print(target_names)
print(n_classes)
print(type(lfw_people))
```

```python
#Initialize PCA and fit the model
Scores = []
components=[]
for i in range(300):
    pca = PCA(n_components=(2 + i))
    principalFeatures = pca.fit_transform(X)
    #Split the dataset into train and test
    X_train, X_test, y_train, y_test = train_test_split(principalFeatures, y, test_size=0.25, shuffle=False)
    X_train_Df = pd.DataFrame(data=X_train)
    y_train_Df = pd.DataFrame(data=y_train, columns=["target"])
    logi = LogisticRegression().fit(X_train_Df, y_train_Df)
    components.append(2 + i)
    Scores.append(logi.score(X_test, y_test))

#print("logi score: " + str(logi.score(X_test, y_test)))
#print(principalDf.shape)
print(principalFeatures.shape)
print(y.shape)
```

```python
fig, ax = plt.subplots()

from matplotlib import ticker

#
ax.scatter(components, Scores, marker='.')

#plt.get_cmap('magma')
ax.set_title("PCA Accuracy with varying components")
ax.set_ylabel("accuracy")
#ax.yaxis.set_minor_formatter(ticker.PercentFormatter(xmax=5))
ax.set_xlabel("# components")
plt.style.use('bmh')
fig.show()

#fig.savefig('yourfilename.png')
#print(len(components)) #300

#find the maximum values of the accuracy..y vector that have less than 75 components

```

```python
#print(Scores)
#print(components)
graph = {"components": components, "Scores": Scores}
graph_df = pd.DataFrame(graph)
#print(graph_df)


data_tuples = list(zip(components,Scores))
#print(data_tuples)
pd.DataFrame(data_tuples, columns=['components','Scores'])

#print(data_tuples)
maximum = 0
for i in range(300):
    if data_tuples[i][1] > maximum and i <= 100:
        maximum = data_tuples[i][1]
        maxPair = data_tuples[i]
        #print(maximum)

print("after using principle component analysis and plotting accuracy vs. numComponents/features, the best performing combination for runtime and accuracy was with:") 
print("components: " + str(maxPair[0]) + " accuracy: " + str(maxPair[1]))
```
